# Get list of adjectives of abstracts containing the word "demonstrator(s)"

code is oriented to the following example at the spaCy documentation page: https://spacy.io/usage/rule-based-matching#example1

```python
# load spacy and set up patterns for matching
import spacy
from spacy.matcher import Matcher

nlp = spacy.load("en_core_web_sm")
matcher = Matcher(nlp.vocab)

patterns = [
    [{'POS':'ADJ'}],
    ]
matcher.add("demo", patterns)


# load abstracts
## Open a text file for reading
arrayAbstracts = []

file_path = "C:\\Users\\fenn\\Nextcloud\\Artikel basale Attribute - article\\get adjectives\\abstracts.txt"  # Replace with the path to your text file
try:
    with open(file_path, 'r', encoding="utf8") as file:
        ## Read the entire file content into a string
      for line in file:
            arrayAbstracts.append(line)
except FileNotFoundError:
    print(f"The file '{file_path}' was not found.")
except Exception as e:
    print(f"An error occurred: {str(e)}")


print(len(arrayAbstracts))

# extract adjectives
listAdjectives = []
for abstract in arrayAbstracts:
    doc = nlp(abstract)
    matches = matcher(doc)
    print(matches)

    for match_id, start, end in matches:
        string_id = nlp.vocab.strings[match_id]  # Get string representation
        span = doc[start:end]  # The matched span
        print(match_id, string_id, start, end, span.text)
        listAdjectives.append(span.text)

# save adjectives
file_path = "C:\\Users\\fenn\\Nextcloud\\Artikel basale Attribute - article\\get adjectives\\listAdjectives.txt"

with open(file_path, 'w', encoding="utf8") as file:
    ## Iterate through the array and write each element to the file
    for adjective in listAdjectives:
        file.write(str(adjective) + '\n')  ## Write each element on a new line
```
