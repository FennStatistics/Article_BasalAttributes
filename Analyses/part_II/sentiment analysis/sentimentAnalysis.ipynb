# Open Text - Sentiment Analysis

## Import modules
for more information regarding the vaderSentiment module see: https://github.com/cjhutto/vaderSentiment
using Vader multi: https://pypi.org/project/vader-multi/

```python
import pandas as pd
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

## Load files

```python
LOADFILE = "C:\\DATEN\\PHD\\Article_BasalAttributes\\Analyses\\part_II\\sentiment analysis\\data_text.csv"


df = pd.read_csv(LOADFILE, delimiter=",", encoding='utf-8')
# rawData = pd.read_csv(LOADFILE, engine='python')
# rawData = pd.read_csv(LOADFILE, engine='python')

#data = pd.read_csv(LOADFILE, encoding='utf-8')

print(len(df))

df['sentimentVadder'] = -99
df['sentimentVadderPolarityScore'] = -99
# iterating the columns
for col in df.columns:
    print(col)
```

## Apply module
Code was partly adjusted from: https://www.geeksforgeeks.org/python-sentiment-analysis-using-vader/

Tresholds were choosen by https://github.com/cjhutto/vaderSentiment and publication

```python
sid_obj = SentimentIntensityAnalyzer()

for i in range(len(df)):
    sentence = df.openQuestion_Ethic_translated[i]
    # print(sentence)
    if pd.isna(sentence) == False :

        sentiment_dict = sid_obj.polarity_scores(sentence)
        # decide sentiment as positive, negative and neutral
        if sentiment_dict['compound'] >= 0.05 :
            print("Positive")
            df.sentimentVadder[i] = "Positive"
            df.sentimentVadderPolarityScore[i] = sentiment_dict['compound']
            
        elif sentiment_dict['compound'] <= - 0.05 :
            print("Negative")
            df.sentimentVadder[i] = "Negative"
            df.sentimentVadderPolarityScore[i] = sentiment_dict['compound']
        else :
            print("Neutral")
            df.sentimentVadder[i] = "Neutral"
            df.sentimentVadderPolarityScore[i] = sentiment_dict['compound']
```

## Save Data Set

```python
df.to_csv("sentiment_text.csv", sep=',')
```

## Check function in more detail

```python
# import SentimentIntensityAnalyzer class
# from vaderSentiment.vaderSentiment module.
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
 
# function to print sentiments
# of the sentence.
def sentiment_scores(sentence):
 
    # Create a SentimentIntensityAnalyzer object.
    sid_obj = SentimentIntensityAnalyzer()
 
    # polarity_scores method of SentimentIntensityAnalyzer
    # object gives a sentiment dictionary.
    # which contains pos, neg, neu, and compound scores.
    sentiment_dict = sid_obj.polarity_scores(sentence)
     
    print("Overall sentiment dictionary is : ", sentiment_dict)
    print("sentence was rated as ", sentiment_dict['neg']*100, "% Negative")
    print("sentence was rated as ", sentiment_dict['neu']*100, "% Neutral")
    print("sentence was rated as ", sentiment_dict['pos']*100, "% Positive")
 
    print("Sentence Overall Rated As", end = " ")
 
    # decide sentiment as positive, negative and neutral
    if sentiment_dict['compound'] >= 0.05 :
        print("Positive")
 
    elif sentiment_dict['compound'] <= - 0.05 :
        print("Negative")
 
    else :
        print("Neutral")

# Driver code
print("\n1st statement : ")
print(df.openQuestion_Ethic_translated[0])
sentence = df.openQuestion_Ethic_translated[0]
# function calling
sentiment_scores(sentence)
 
print("\n2nd Statement :")
print(df.openQuestion_Ethic_translated[1])
sentence = df.openQuestion_Ethic_translated[1]
sentiment_scores(sentence)
```
